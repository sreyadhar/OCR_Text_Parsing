```python
# Created by Sreya Dhar
# Date : 09/14/2021
# Python program to read json file
import json
from pprint import pprint
from itertools import chain

# Opening JSON file
filepath_json = 'testimage1.json'
f = open(filepath_json,)
   
# returns JSON object as a dictionary
data = json.load(f)
   
# Iterating through the json list
val = []
for i in data['Blocks'][1:]:
    # print(i)
    for k,v in i.items():
      if k =='Text':
        val.append(v)

# pprint(val)
leng = round(len(val)/2)-1
data_ = val[:leng]

# Closing file
f.close()

val1 = []
for i in data_:
  i=i.split()
  val1.append(i)
# print(val1)
val2 = []
for i in val1:
  for j in i:
    lim = '/,.'
    for k in j:
      if k in lim:
        j = j.replace(k, "")
      j = j.replace('top', "00")
      j = j.replace('too', "00")
      j = j.replace('100', "00")
      j = j.replace('R', '2')
      j = j.replace('in', 'A')
    if j!='':
      val2.append(j)
  
```

```python
chunks = [val2[x:x+12] for x in range(0, len(val2), 12)]
del chunks[2][4]
chunks[2][3]='10'
val3 = []
for i in chunks:
  for j in i:
    val3.append(j)
chunks = [val3[x:x+12] for x in range(0, len(val3), 12)]
chunks[3].insert(8, 'A')
val4 = []
for i in chunks:
  for j in i:
    val4.append(j)
chunks = [val4[x:x+12] for x in range(0, len(val4), 12)]

for i,x in enumerate(chunks[1:]):
  for j,a in enumerate(x):
    if x[5]!='21':
      x[5]='21'
    if x[7]!='TC':
      x[7]='TC'
    if x[11]!='1':
      x[11]='1'

chunks[0][6], chunks[2][6],chunks[3][9] ='95940', '95927', '7013'

for i in chunks:
  del i[0]
  z=''.join(i[1:4])
  del i[1:4]
  i.insert(1,z)
  i.insert(3,'')

for j in chunks:
  for i in range(3):
    j.insert(6,'')
chunks_1=chunks
# for i in chunks_1:
#   print(i)

testimage1 = {'testimage1':chunks_1}
pprint(testimage1)
```
